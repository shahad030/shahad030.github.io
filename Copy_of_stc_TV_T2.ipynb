```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go

# ============================================================================
# STEP 1: PREPARE DATA (date_ is already the index)
# ============================================================================

# Your dataframe already has date_ as index
df = dataframe.copy()

# Ensure index is datetime
df.index = pd.to_datetime(df.index)

# Sort by date
df = df.sort_index()

print("="*70)
print("DATA OVERVIEW")
print("="*70)
print(f"Total samples: {len(df)}")
print(f"Date range: {df.index.min().date()} to {df.index.max().date()}")
print(f"Watch time - Mean: {df['Total_watch_time_in_houres'].mean():.2f} hours")
print(f"Watch time - Range: {df['Total_watch_time_in_houres'].min():.2f} - {df['Total_watch_time_in_houres'].max():.2f} hours")

# ============================================================================
# STEP 2: SPLIT DATA FOR MODEL VALIDATION
# ============================================================================

n = len(df)
train_size = int(n * 0.70)  # 70% for training
val_size = int(n * 0.15)     # 15% for validation
test_size = n - train_size - val_size  # 15% for testing

train_df = df.iloc[:train_size].copy()
val_df = df.iloc[train_size:train_size+val_size].copy()
test_df = df.iloc[train_size+val_size:].copy()

print(f"\nData Split:")
print(f"  Train: {len(train_df)} days ({train_df.index.min().date()} to {train_df.index.max().date()})")
print(f"  Validation: {len(val_df)} days ({val_df.index.min().date()} to {val_df.index.max().date()})")
print(f"  Test: {len(test_df)} days ({test_df.index.min().date()} to {test_df.index.max().date()})")

# ============================================================================
# STEP 3: TEST MULTIPLE MODELS
# ============================================================================

def calculate_mae(actual, predicted):
    return np.mean(np.abs(actual - predicted))

def calculate_rmse(actual, predicted):
    return np.sqrt(np.mean((actual - predicted)**2))

train_ts = train_df['Total_watch_time_in_houres'].values
val_ts = val_df['Total_watch_time_in_houres'].values

print(f"\n{'='*70}")
print("MODEL EVALUATION ON VALIDATION SET")
print("="*70)

models_results = {}

# Model 1: Moving Average
window = 7
ma_predictions = []
for i in range(len(val_ts)):
    if i == 0:
        recent = train_ts[-window:]
    else:
        recent_train = train_ts[-window+i:] if window > i else []
        recent_val = val_ts[:i]
        recent = np.concatenate([recent_train, recent_val])[-window:]
    ma_predictions.append(recent.mean())

mae_ma = calculate_mae(val_ts, np.array(ma_predictions))
models_results['Moving Average (7-day)'] = mae_ma
print(f"1. Moving Average (7-day): MAE = {mae_ma:.2f} hours")

# Model 2: Linear Trend
x = np.arange(len(train_ts))
y = train_ts
x_mean = np.mean(x)
y_mean = np.mean(y)
slope = np.sum((x - x_mean) * (y - y_mean)) / np.sum((x - x_mean)**2)
intercept = y_mean - slope * x_mean

x_val = np.arange(len(train_ts), len(train_ts) + len(val_ts))
trend_predictions = slope * x_val + intercept
trend_predictions = np.clip(trend_predictions, 0, None)

mae_trend = calculate_mae(val_ts, trend_predictions)
models_results['Linear Trend'] = mae_trend
print(f"2. Linear Trend: MAE = {mae_trend:.2f} hours (slope = {slope:.2f} hrs/day)")

# Model 3: Naive (last value)
naive_predictions = np.full(len(val_ts), train_ts[-1])
mae_naive = calculate_mae(val_ts, naive_predictions)
models_results['Naive'] = mae_naive
print(f"3. Naive (last value): MAE = {mae_naive:.2f} hours")

# Select best model
best_model = min(models_results, key=models_results.get)
print(f"\n Best Model: {best_model} (MAE = {models_results[best_model]:.2f} hours)")

# ============================================================================
# STEP 4: MAKE PREDICTIONS FOR NEXT 2 MONTHS
# ============================================================================

print(f"\n{'='*70}")
print("GENERATING PREDICTIONS FOR NEXT 2 MONTHS")
print("="*70)

# Use all data to train final model
full_ts = df['Total_watch_time_in_houres'].values

# Calculate linear trend on full data
x = np.arange(len(full_ts))
y = full_ts
x_mean = np.mean(x)
y_mean = np.mean(y)
slope = np.sum((x - x_mean) * (y - y_mean)) / np.sum((x - x_mean)**2)
intercept = y_mean - slope * x_mean

print(f"Model: Linear Trend")
print(f"  Slope: {slope:.4f} hours/day")
print(f"  This means watch time is {'decreasing' if slope < 0 else 'increasing'} by {abs(slope):.2f} hours per day")

# Generate predictions for next 61 days (May + June)
last_date = df.index.max()
future_dates = pd.date_range(start=last_date + pd.Timedelta(days=1), periods=61, freq='D')

x_future = np.arange(len(full_ts), len(full_ts) + 61)
future_predictions_values = slope * x_future + intercept
future_predictions_values = np.clip(future_predictions_values, 0, None)  # No negatives

# Create predictions dataframe with date as index
future_predictions = pd.DataFrame({
    'Total_watch_time_in_houres': future_predictions_values
}, index=future_dates)

# Monthly summary
may_pred = future_predictions[future_predictions.index.month == 5]
june_pred = future_predictions[future_predictions.index.month == 6]

print(f"\n MAY 2018:")
print(f"  Average daily: {may_pred['Total_watch_time_in_houres'].mean():.2f} hours")
print(f"  Total: {may_pred['Total_watch_time_in_houres'].sum():.2f} hours")

print(f"\n JUNE 2018:")
print(f"  Average daily: {june_pred['Total_watch_time_in_houres'].mean():.2f} hours")
print(f"  Total: {june_pred['Total_watch_time_in_houres'].sum():.2f} hours")

print(f"\nFirst 5 predictions:")
print(future_predictions.head())

# ============================================================================
# STEP 5: VISUALIZE RESULTS
# ============================================================================

print(f"\n{'='*70}")
print("CREATING VISUALIZATION")
print("="*70)

# Combine historical and predictions
combined_df = pd.concat([df, future_predictions])

# Reset index to make date a column for plotly
combined_plot = combined_df.reset_index()
combined_plot.columns = ['date_', 'Total_watch_time_in_houres']

# Create the plot using plotly (your preferred style)
fig = px.line(combined_plot, x='date_', y='Total_watch_time_in_houres',
              title='Watch Time: Historical Data + 2-Month Forecast')

# Add a vertical line to show where predictions start
fig.add_vline(x=df.index.max(), # Use Timestamp directly for the line
              line_dash="dash",
              line_color="red")

# Add a separate annotation for the "Forecast Starts" text
fig.add_annotation(x=df.index.max(), # Position the annotation at the end of historical data
                   y=1.02, # Slightly above the top of the plot (normalized coordinates)
                   yref="paper", # y is in paper coordinates (0 to 1)
                   text="Forecast Starts",
                   showarrow=False,
                   font=dict(color="red"),
                   bgcolor="rgba(255,255,255,0.8)",
                   align="center")

# Customize layout
fig.update_layout(
    xaxis_title='Date',
    yaxis_title='Total Watch Time (hours)',
    hovermode='x unified',
    showlegend=True
)

# Show the plot
fig.show()

# ============================================================================
# STEP 6: SAVE PREDICTIONS
# ============================================================================

# Save predictions to Excel (reset index to include date column)
future_predictions.reset_index().to_excel('watch_time_predictions_May_June_2018.xlsx', index=False)
print("\n Predictions saved to 'watch_time_predictions_May_June_2018.xlsx'")

# Also save combined data
combined_df.reset_index().to_excel('watch_time_historical_and_predictions.xlsx', index=False)
print(" Combined data saved to 'watch_time_historical_and_predictions.xlsx'")

print(f"\n{'='*70}")
print("SUMMARY")
print("="*70)
print(f" Historical data: {len(df)} days")
print(f" Predictions: {len(future_predictions)} days (May 1 - June 30, 2018)")
print(f" Best model: {best_model}")
print(f" Average predicted watch time: {future_predictions['Total_watch_time_in_houres'].mean():.2f} hours/day")
print(f" Trend: {'Declining' if slope < 0 else 'Increasing'} by {abs(slope):.2f} hours per day")
```
